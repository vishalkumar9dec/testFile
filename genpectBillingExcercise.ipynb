## Importing the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Reading the data and skipping the first three rows as it contains the source information.

```python
df_billing = pd.read_csv('billings_europe.csv',skiprows=3)
```

```python
df_billing.shape
```

### Dropping the rows and columns whose all values are null

```python
df_billing.dropna(how='all',inplace=True)
df_billing.dropna(axis=1,how='all',inplace=True)
df_billing.shape
```

```python
df_billing=df_billing.set_index('1')
```

```python
df_nonNull = df_billing.dropna(how='all')
```

```python
df_nonNull.shape
```

### Segregatting the data on the basis of requirement. As we have to select the billing 1st Jan 2016 till date

```python
updated_data = df_nonNull.loc['1-Jan-16':]
```

```python
updated_data.shape
```

```python
updated_data=updated_data.fillna(0)
```

```python
updated_data.loc['Value'] = updated_data.sum()
```

```python
updated_data['date']= updated_data.index
```

```python
updated_data
```

```python
updated_data['date'].nunique()
```

```python
test_ud = updated_data.copy()
```

```python
test_ud = test_ud.transpose()
```

```python
test_ud.head()
```

```python
data_df = pd.DataFrame()
for i in range(3):
    df = pd.DataFrame(df_billing.iloc[i])
    #df.drop('1', axis=0, inplace=True)
    data_df = pd.concat([data_df,df],axis=1,sort=False)

data_df.columns = ['Segment','Type','Subtype']
```

```python
data_df.head()
```

```python
data_df[['Segment','Period']] = data_df['Segment'].str.split('-',expand=True)
```

```python
data_df = data_df[['Segment','Period','Type','Subtype']]
data_df.head()
```

```python
data_df=data_df.fillna(method='ffill')
```

```python
data_df.head()
```

```python
data_df['Value'] = updated_data.loc['Value']
```

```python
data_df['Value'] = data_df['Value'].map(lambda x : '%.5f' %x)
```

```python
data_df.head()
```

```python
#Keeping a backup of the data
final_data = data_df.copy()
```

```python
data_df['Value'] = data_df['Value'].astype('float')
data_df.head()
```

```python
Country_billing_data = data_df[data_df['Type']=="Countries"]
```

```python
Country_billing_data.shape
```

```python
Country_billing_data.info()
```

```python
Country_billing = Country_billing_data.groupby('Subtype').sum()
```

```python
Country_billing=Country_billing.rename(columns={'Value':'Billing'})
Country_billing
```

```python
Period_billing = data_df.groupby('Period').sum()
Period_billing = Period_billing.rename(columns={'Value': 'Billing'})
Period_billing
```
